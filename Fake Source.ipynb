```python
import random
import numpy as np
import pandas as pd
from faker import Faker

fake = Faker()

data1 = []
for i in range(500):
  data1.append({
      "tenant_id": f"T{i:03}",
      "tenant_name": fake.company(),
      "tenant_email": fake.email(),
      "tenant_phone": fake.phone_number(),
      "tenant_address": fake.address(),
      "industry": random.choice(["FMCG","Energy","Manufacturing","Utilities","Tech"]),
      "location":random.choice(['Cikarang','Bogor','Bandung','Jakarta','Surabaya'])
  })

data1 = pd.DataFrame(data1)
data1.to_csv("Tenant.csv", index=False)


data2 = []
for i in range(200):
  for day in range(1, 6):
    data2.append({
        "tenant_id": f"T{i:03}",
        "date": f"2024-01-0{day}",
        "gas_m3": random.randint(10, 1000),
        "water_m3": random.randint(10, 100),
        "electricity_kwh": random.randint(500, 5000)
  })

data2 = pd.DataFrame(data2)
data2.to_csv("Utilities.csv", index=False)



data3 = []
for i in range(700):
  data3.append({
      "transaction_id":f"TRX{i}{day}",
      "tenant_id": f"T{i:03}",
      "date": f"2024-01-0{day}",
      "amount": random.randint(10000, 500000),
      "tenor" : random.choice([3,6,12]),
      "addres": random.choice(['Cikarang','Bogor','Bandung','Jakarta','Surabaya']),
      "status": random.choice(['No-Default','Default'])
  })

data3 = pd.DataFrame(data3)
data3.to_csv("Transactions.csv", index=False)
```
